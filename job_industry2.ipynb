```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import seaborn as sns
df = pd.read_csv('trainData.csv')
df2 = pd.read_csv('testData.csv')
df.head()
```

```python
df.isnull().sum()
```

Train Data Preprocessing

```python
df= df.drop(['Id','jobid','postdate','numberofpositions','payrate'],axis =1)
df['company'].fillna(df['company'].mode()[0],inplace = True)
df['industry'].fillna(df['industry'].mode()[0],inplace = True)
df['skills'] = np.where(df['skills'] == 0.0,np.nan,df['skills'])
df['skills'] = df.groupby('industry')['skills'].transform(lambda x:x.fillna(x.mode()[0]))

df['experience'] = df['experience'].str.split(" ", n = 1,expand = True)
df['experience'] = df['experience'].replace({'Missing':0,'Not':0})

df['jobdescription'] = df['jobdescription'].str.split(" ", n = 8,expand = True)[8]
df= pd.get_dummies(df,drop_first=True)
```

Test Data Preprocessing

```python
df2= df2.drop(['Id','jobid','postdate','numberofpositions','payrate'],axis =1)
df2['company'].fillna(df2['company'].mode()[0],inplace = True)
#df2['industry'].fillna(df2['industry'].mode()[0],inplace = True)
df2['skills'] = np.where(df2['skills'] == 0.0,np.nan,df2['skills'])
df2['skills'] = df2['skills'].transform(lambda x:x.fillna(x.mode()[0]))

df2['experience'] = df2['experience'].str.split(" ", n = 1,expand = True)
df2['experience'] = df2['experience'].replace({'Missing':0,'Not':0})

df2['jobdescription'] = df2['jobdescription'].str.split(" ", n = 8,expand = True)[8]
df2= pd.get_dummies(df2, columns=["skills" ], prefix=["skills"],drop_first=True)
df2.isnull().sum()
```
